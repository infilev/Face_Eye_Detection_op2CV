### COMBINING FACE AND EYE DETECTION

```python
import numpy as np
import cv2
face_classifier = cv2.CascadeClassifier("Haarcascades\\haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("Haarcascades\\haarcascade_eye.xml")
```

```python
img = cv2.imread("C:\\Users\\Anubhav\\Desktop\\Data Science\\OpenCV\\Face and Eye Detection\\image_examples\\Trump.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# stores the top left coordinate and bottom right coordinate (detectMultiScale : extracts features)
faces = face_classifier.detectMultiScale(gray, 1.3, 5)

#when face is not detected, face_classifier returns and empty tuple

if len(faces) == 0:
    print("NO FACES FOUND")
    

#  WE will iterate through our faces array and draw a rectangle over each face in first loop (outer one)
    
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (127, 0 ,255), 2)
    cv2.imshow('img',img)
    cv2.waitKey(0)

    #crop that face and getting image details in roi_gray and roi_color
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]
    eyes = eye_classifier.detectMultiScale(roi_gray)
    
    #width and height of eyes
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (255, 0 ,255), 2)
        cv2.imshow('img',img)
        if cv2.waitKey(0) & 0xFF==ord('x'):
                                      break
        
cv2.destroyAllWindows()

#after running press "ENTER" to see the eye rectangle
```
